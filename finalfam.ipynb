# 4. Proportion of Residential/Commercial/Other EC Houses 

```python
d41 = pd.DataFrame(comm[['SURVEY AREA TYPE', 'SURVEY ID']].groupby(['SURVEY AREA TYPE']).count().reset_index())
d41.columns = ['SURVEY AREA TYPE', 'Count_of_commercial']

d42 = pd.DataFrame(oth[['SURVEY AREA TYPE', 'SURVEY ID']].groupby(['SURVEY AREA TYPE']).count().reset_index(drop = True))
d42.columns = ['Count_of_others']

d43 = pd.DataFrame(cum[['SURVEY AREA TYPE', 'SURVEY ID']].groupby(['SURVEY AREA TYPE']).count().reset_index(drop = True))
d43.columns = ['Count_of_cum']

matrix1 = pd.concat([d41, d42, d43], axis = 1)
```

```python
matrix1 = matrix1.pivot_table(index='SURVEY AREA TYPE',
               margins=True,
               margins_name='total',  # defaults to 'All'
               aggfunc=sum).reset_index()

matrix1['total'] = matrix1.sum(axis = 1)
matrix1.set_index('SURVEY AREA TYPE', inplace = True)
```

```python
display(matrix1.style.set_properties(**{'background-color': 'black',
                           'color': 'lawngreen',
                           'border-color': 'white'}))
```
